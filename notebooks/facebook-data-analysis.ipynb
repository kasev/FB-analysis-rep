```python
### we extract all messages as a list of words
### from the messages, we take only first 160 characters and cut off the last words as they are often uncomplete.

to_replace = {"COUNTRY LIFE BIO": "CountryLifeBio", "COUNTRY LIFE": "CountryLifeBio", "Country Life Bio" : "CountryLifeBio", "Country Life BIO" : "CountryLifeBio", "Country Life" : "CountryLifeBio", "Country life bio": "CountryLifeBio", "Country Life bio": "CountryLifeBio", "Country life bio": "CountryLifeBio", "Country life": "CountryLifeBio"}
def replace_all(text, to_replace):
    for i, j in to_replace.items():
        text = text.replace(i, j)
    return text

def words_only(df):
    words_list=[]
    for m in df["message"]:
        try:
            separated_words=replace_all(m[:159], to_replace).split(" ")
            separated_words=separated_words[:-1]
            ### separated_words.append("...") - if we want separated
        except:
            separated_words=[]
        for word in separated_words:
                word=word.strip()
                ### word=word.lower()
                words_list.append(word)
    return words_list
```

```python
### Here you put the INPUT
words = words_only(least_likes)
###


words_joined = " ".join(words)

tokenizer.setText(words_joined)
t = 0
output = []

while tokenizer.nextSentence(forms, tokens):
      tagger.tag(forms,lemmas)
      
      for i in range (len(lemmas)):
        form = forms[i]
        output.append([form, lemmas[i].lemma, lemmas[i].tag])
        
output_df = pd.DataFrame(output, columns=("form", "lemma", "TAG"))


### interested only in a subselection of words
output_df = output_df[output_df["TAG"].str.startswith(("N","V","A"))]


### let make a list from it
text_lemmata  = output_df["lemma"].tolist()

text_lemmata_short = []
for lemma in text_lemmata:
    ### we will ignore everthing but the words themselves
    text_lemmata_short.append(lemma.partition("_")[0].replace("-1",""))

    
# and finally derive frequency distribution of the lemmata
text_lemmata_fd = nltk.FreqDist([word for word in text_lemmata_short])
text_lemmata_fd.most_common(50)

# transforming it into a dataframe
lemmata_df = pd.DataFrame(text_lemmata_fd.most_common(30), columns=["lemma", "count"])
```
